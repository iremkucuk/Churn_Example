```python
def check_outlier(dataframe,col_name):
    low_limit,up_limit=outlier_treshold(dataframe,col_name)
    if dataframe[(dataframe[col_name]< low_limit)|(dataframe[col_name]> up_limit)].any(axis=None):
        return True
    else:
        return False 
```

```python
def outlier_treshold(dataframe,col_name,q1=0.25,q3=0.75):
    quartile1=dataframe[col_name].quantile(q1)
    quartile3=dataframe[col_name].quantile(q3)
    interquartile=quartile3-quartile1
    up_limit=quartile3+1.5*interquartile
    low_limit=quartile1-1.5*interquartile
    return low_limit,up_limit
```
